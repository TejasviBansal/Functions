## Theory Questions

1. What is the difference between a function and a method in Python?

   -> **Function**

      1. A function is a block of reusable code that performs a specific task.
      2. Defined using the def keyword.
      3. Can exist independently (outside of any class).
      4. Called directly using its name.
      5. Example:

             def greet(name):
             return f"Hello, {name}!"
             print(greet("Alice"))   # Hello, Alice!


   -> **Method**

      1. A method is also a function, but it is associated with an object.
      2. Defined inside a class.
      3. It always takes the instance (self) as its first argument (for instance methods).
      4. Called using the dot notation on an object.
      5. Example:

             class Person:
                def greet(self):
                    return "Hello, I am a person."
             p = Person()
             print(p.greet())   # Hello, I am a person.

2. Explain the concept of function arguments and parameters in Python.

   -> **Parameters**

      1. Placeholders in the function definition.
      2. They act like variables that receive values when the function is called.
      3. Defined inside the parentheses () of a function.
      4. Example:

             def greet(name):   # 'name' is a parameter
                 print(f"Hello, {name}!")
            
      **Arguments**

      1. Actual values you pass to the function when calling it.
      2. They fill in the placeholders (parameters).
      3. Example:

             greet("Alice")   # "Alice" is an argument

3. What are the different ways to define and call a function in Python?

   -> 1. **Simple Function** (No Parameters)
          
        def greet():
            print("Hello, World!")

      Calling:
        
       greet()

      2. **Function with Parameters**

        def greet(name):
            print(f"Hello, {name}!")

      Calling:

       greet("Alice")

      3. **Function with Default Parameters**

       def greet(name="Guest"):
            print(f"Hello, {name}!")
      Calling:

       greet()          # Hello, Guest
       greet("Bob")     # Hello, Bob

     4. **Function with Return Value**

       def add(a, b):
           return a + b
     Calling:

       result = add(5, 3)  # 8

     5. **Function with Variable-Length Arguments (*args)**

        Used when you don’t know how many arguments will be passed.

       def add_numbers(*args):
           return sum(args)

      Calling:

       print(add_numbers(1, 2, 3, 4))   # 10

     6. **Function with Keyword Arguments (**kwargs)**

        Used to accept key-value pairs.

       def introduce(**kwargs):
           for key, value in kwargs.items():
               print(f"{key}: {value}")
      Calling:

       introduce(name="Alice", age=21, course="Python")

     7. **Lambda Function (Anonymous Function)**

        Small one-line function using lambda.

       square = lambda x: x * x
      Calling:

       print(square(5))   # 25

     8. **Recursive Function**

        A function that calls itself.

       def factorial(n):
           if n == 0:
               return 1
           return n * factorial(n-1)

     Calling:

       print(factorial(5))   # 120

6. Explain the concept of generators in Python and how they are defined.

   1. A generator is a special type of iterator that allows you to generate values on the fly instead of storing them in memory.
   2. They are memory-efficient because they don’t create the entire sequence at once. Instead, they produce one item at a time when needed.
   3. Generators are created using:

        1. Generator functions (with yield)

        2. Generator expressions (like list comprehensions, but with parentheses)

    4. A generator function is just like a normal function, but instead of return, it uses yield.
    5. Example:

           def count_up_to(n):
               i = 1
               while i <= n:
                   yield i   # yields one value at            a time
                   i += 1

           gen = count_up_to(5)

           print(next(gen))  # 1
           print(next(gen))  # 2
           print(next(gen))  # 3

7. What are the advantages of using generators over regular functions?

   1. **Memory Efficiency**

        1. A regular function (or list comprehension) often creates and stores the entire result in memory.

        2. A generator produces one value at a time and does not store the whole sequence.

      Useful for large datasets or infinite sequences.

   2. **Lazy Evaluation (On-Demand Computation)**

        1. Generators compute values only when needed (next() or in a loop).

        2. Saves time if you don’t need all results at once.

   3. **Infinite Sequences Possible**

        1. Regular functions cannot handle infinite series (e.g., natural numbers) because they would try to create an endless list.

        2. Generators can represent infinite sequences safely.

   4. **Improved Performance**

        1. Since values are generated one by one, you avoid unnecessary computations when not all values are needed.

        2. This makes loops over large datasets faster and more efficient.

   5. **Cleaner and More Readable Code**

        1. Generators simplify code compared to manually creating and managing iterators with __iter__() and __next__().

8. What is a lambda function in Python and when is it typically used?

   -> A lambda function in Python is a small, anonymous function (i.e., it has no name).

     1. Defined using the keyword lambda instead of def.

     2. Can take any number of arguments but must have only one expression.

     3. That expression is evaluated and returned automatically.

   General syntax:

       lambda arguments: expression


   Example:

       square = lambda x: x * x
       print(square(5))   # Output: 25

   Lambda functions are used when you need a short, throwaway function — often passed as an argument to higher-order functions.

9. Explain the purpose and usage of the `map()` function in Python.

   -> The map() function applies a given function to each item in an iterable (like a list, tuple, or set) and returns a map object (which is an iterator).

   Synatx: map(function, iterable)

   map() can take multiple iterables. The function should accept as many arguments as the number of iterables.

   Why use map()?

     1. Makes code shorter and cleaner

     2. Avoids manual loops (for loop)

     3. Works seamlessly with lambda functions for quick transformations

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

     -> **map()**

     Purpose: Transform each element of an iterable using a function.

      1. Applies the function to every element.

      2. Returns a map object (an iterator).

      3. Example:

              nums = [1, 2, 3, 4]
              result = map(lambda x: x * 2, nums)
              print(list(result))   # [2, 4, 6, 8]

      **filter()**

      Purpose: Select elements from an iterable based on a condition (True/False).

       1. Keeps only elements where the function returns True.

       2. Returns a filter object (an iterator).

       3. Example:

              nums = [1, 2, 3, 4, 5, 6]
              result = filter(lambda x: x % 2 == 0, nums)
              print(list(result))   # [2, 4, 6]

      **reduce() (from functools)**

      Purpose: Repeatedly apply a function to elements of an iterable, reducing it to a single value.

       1. Function must take two arguments.

       2. Returns a single cumulative result.

       3. Example:

              from functools import reduce

              nums = [1, 2, 3, 4, 5]
              result = reduce(lambda x, y: x + y, nums)
              print(result)   # 15

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    """Check if a number is prime."""
    if n <= 1:
        return False   # 0 and 1 are not prime
    for i in range(2, int(n**0.5) + 1):  # check divisibility up to sqrt(n)
        if n % i == 0:
            return False
    return True

# Print primes from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # total number of terms to generate
        self.count = 0          # keeps track of how many terms generated
        self.a, self.b = 0, 1   # first two Fibonacci numbers

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n_terms:
            if self.count == 0:
                self.count += 1
                return self.a
            elif self.count == 1:
                self.count += 1
                return self.b
            else:
                self.a, self.b = self.b, self.a + self.b
                self.count += 1
                return self.b
        else:
            raise StopIteration   # stops iteration when done

fib = Fibonacci(10)   # generate first 10 Fibonacci numbers
for num in fib:
    print(num, end=" ")
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filename):
    """Generator function that yields lines from a file one by one."""
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()   # remove newline characters

for line in read_file_line_by_line("sample.txt"):
    print(line)
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
# List of tuples
data = [(1, 4), (3, 1), (5, 9), (2, 6), (7, 2)]

# Sort by the second element of each tuple
sorted_data = sorted(data, key=lambda x: x[1])

print("Original list:", data)
print("Sorted list:", sorted_data)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# List of temperatures in Celsius
celsius = [0, 20, 37, 100, -10]

# Convert to Fahrenheit using map() and lambda
fahrenheit = list(map(lambda c: (9/5) * c + 32, celsius))

print("Celsius:    ", celsius)
print("Fahrenheit: ", fahrenheit)
```
